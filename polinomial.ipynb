```python
drp=["ID","Date House was Sold","Zipcode","Longitude","Latitude"]
df=df.drop(columns=drp)
df.head()
```

```python
df.columns
```

```python
drp1=["No of Times Visited","Renovated Year","Condition of the House"]
df=df.drop(columns=drp1)
df.shape, df.head()
```

```python
df.isna().sum()
```

```python
df["Sale Price"].isna().sum()
```

```python
df = df.dropna(subset=["Sale Price"])
df.isna().sum()
```

```python
num_cols = df.select_dtypes(include="number").columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())
df.isna().sum()
```

```python
drp2 = ["Waterfront View"]
df = df.drop(columns=drp2)
df.select_dtypes(exclude="number").columns
df.shape, df.head()
```
